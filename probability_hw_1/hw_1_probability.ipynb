Для каждого распределения поэкспериментируйте с выбором параметров и изобразите их на одном графике, не забывая про легенду и подписи осей.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
# Python 2 and 3 compatibility
# pip install future
from __future__ import (absolute_import, division,
                        print_function, unicode_literals)
# отключим предупреждения Anaconda
import warnings
warnings.simplefilter('ignore')

# будем отображать графики прямо в jupyter'e
%pylab inline
#графики в svg выглядят более четкими
%config InlineBackend.figure_format = 'svg' 

#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 6,3
import pandas as pd
import seaborn as sns
```

```python
# (a) Probability mass function
mu_1 = 5
rv_1 = sts.poisson(mu_1)
mu_2 = 6
rv_2 = sts.poisson(mu_2)
mu_3 = 7
rv_3 = sts.poisson(mu_3)
x = np.linspace(0,20,21)
plt.plot(x,rv_1.pmf(x),'o')
plt.plot(x,rv_2.pmf(x),'o')
plt.plot(x,rv_3.pmf(x),'o')
plt.legend ([r'$\mu=%i$'% mu_1,r'$\mu=%i$'% mu_2,r'$\mu=%i$'% mu_3])
plt.xlabel('$x$')
plt.ylabel(r'$p(x|\mu)$')
plt.title('Poisson Distribution')
```

```python
# (a) Cumulative density function
mu = 4
poisson_rv = sts.poisson(mu)
x = np.linspace(0,20,21)
cdf = poisson_rv.cdf(x)
plt.plot(x, cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.title('Function Distribution')
```

```python
# (b) Probability density function

expon_rv_1 = sts.expon(scale = 10)
expon_rv_2 = sts.expon(scale = 15)
expon_rv_3 = sts.expon(scale = 20)
x = np.linspace(0,20,21)
plt.plot(x,expon_rv_1.pdf(x))
plt.plot(x,expon_rv_2.pdf(x))
plt.plot(x,expon_rv_3.pdf(x))

plt.legend ([r'$scale=%i$'% 10,r'$scale=%i$'% 15,r'$scale=%i$'% 20])
plt.xlabel('$x$')
plt.ylabel(r'$f(x)$')
plt.title('Expon Distribution')
```

```python
# (b) Cumulative density function
expon_rv = sts.expon()
x = np.linspace(0,20,21)
cdf = expon_rv.cdf(x)
plt.plot(x, cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.title('Function Distribution')
```

```python
# (c) A Planck discrete exponential random variable
lambda_1 = 0.51
rv_plank_1 = sts.planck(lambda_1)
lambda_2 = 0.51/2
rv_plank_2 = sts.planck(lambda_2)
lambda_3 = 0.51/4
rv_plank_3 = sts.planck(lambda_3)
x = np.linspace(0,20,21)
plt.plot(x,rv_plank_1.pmf(x),'o')
plt.plot(x,rv_plank_2.pmf(x),'o')
plt.plot(x,rv_plank_3.pmf(x),'o')

plt.legend ([r'$\lambda=%f$'% lambda_1,r'$\lambda=%f$'% lambda_2,r'$\lambda=%f$'% lambda_3,])
plt.xlabel('$x$')
plt.ylabel(r'$f(x)$')
plt.title('Plank Distribution')
```

```python
# (c) Cumulative density function
lambda_ = 0.51
plank_rv = sts.planck(lambda_)
x = np.linspace(0,20,21)
cdf = plank_rv.cdf(x)
plt.plot(x, cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.title('Function Distribution')
```
