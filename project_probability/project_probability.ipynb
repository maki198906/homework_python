```python
# Проверим взаимосвязь между количеством проведенных лет и сатисфакцией
```

```python
work_3 = data[data['time_spend_company'] == 3]['satisfaction_level']

work_7 = data[data['time_spend_company'] == 7]['satisfaction_level']

#print(work_3.var(),work_7.var()) гомогенность дисперсий -> можно применить t-test
# Выдвинем гипотезу , что M3=M4 (H0)
# Применим t -test
stats.ttest_ind(work_3,work_7)
# т.к. p-value >0.05 , то оснований отвергать гипотезу Н0 нет
```

```python
# Проверить несколько гипотез для данных, чтобы понять почему сотрудники покидают компанию
```

```python
new_data = data.drop(['Количество работников'],axis = 1)
new_data.head()
```

```python
# разделим дата сет на тех, кто ушел и тех, кто остался
```

```python
who_gone = new_data[(new_data['left'] == 1)]
```

```python
who_stay = new_data[(new_data['left'] == 0)]
```

```python
# сравним с помощью t-test уровень сатисфакции для тех, кто ушел и кто все еще работает
# Гипотеза: если у нас H0: Мstay = Мgone,то связь между сатисфакцией и уходом можно отвергнуть
```

```python
gone_stf = who_gone['satisfaction_level']
```

```python
stay_stf = who_stay['satisfaction_level']
```

```python
stats.ttest_ind(stay_stf,gone_stf)
# т.к. p-value <0.05 , то гипотезу, что сотрудники уходят от неудовлетворенности отвергать не стоит
```

```python
# сравним с помощью t-test уровень последней оценки работы для тех, кто ушел и кто все еще работает
# Гипотеза: если у нас H0: Мstay = Мgone,то связь между последней оценки работы и уходом можно отвергнуть
```

```python
gone_evl = who_gone['last_evaluation']
```

```python
stay_evl = who_stay['last_evaluation']
```

```python
stats.ttest_ind(stay_evl,gone_evl)
# т.к. p-value >0.05 , то гипотезу, что сотрудники уходят из-за последней оценки работы можно отвергнуть
```

```python
# сравним с помощью t-test среднее количество часов в месяц работы для тех, кто ушел и кто все еще работает
# предположим, что среднее различие в количестве часов работы в месяц 
# у тех кто работает много и получает мало и тех, кто работает мало и получает много одинаково
# H0: Мstay = Мgone
```

```python
k = who_gone[['average_montly_hours','salary']]
salary_low = k[k['salary'] == 'low']
```

```python
j = who_stay[['average_montly_hours','salary']]
salary_high = j[j['salary'] == 'high']
```

```python
stats.ttest_ind(salary_high['average_montly_hours'],salary_low['average_montly_hours'])
# т.к. p-value <0.05, то гипотезу, что сотрудники уходят из-за различия в количестве часов работы в месяц
# можно отвергнуть
```

```python
# можно проверить гипотезу , что среднее количесвто часов в месяц может отличаться для тех, кто ушел и тех, кто нет
```

```python
b = who_gone[['average_montly_hours','salary']]
salary_high_gone = b[b['salary'] == 'high']
```

```python
stats.ttest_ind(salary_high['average_montly_hours'],salary_high_gone['average_montly_hours'])
# но т.к. p-value > .05, то отвергать гипотезу , что H0: Мstay = Мgone не стоит, т.е.
# кол-во часов может быть одинаковым и не являться причиной ухода
```
