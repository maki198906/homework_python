#### Проверить найденные пары t-тестом

```python
# your solution is here
# гипотеза - совадает ли среднее
df_1 = df[df['mean radius'] > 14]['mean perimeter']
df_2 = df[df['mean radius'] < 14]['mean perimeter']
print(stats.ttest_ind(df_1,df_2))
# mean radius имеет значение для mean perimeter

df_3 = df[df['mean texture'] > 19]['mean smoothness']
df_4 = df[df['mean texture'] < 19]['mean smoothness']
print(stats.ttest_ind(df_3,df_4))
# mean texture не имеет значение для mean texture mean smoothness
```

#### Проверить руками гипотезу $H_0: M_{mean\ compactness} = 0$

#### Нарисовать распределение признака worst texture с помощью
- диаграммы
- qq-plot
- эмпирическая ф.р.

```python
# your solution is here
df[['worst texture']].describe()
```

```python
# для четкости
%config InlineBackend.figure_format = 'svg'
x = df['worst texture']
plt.figure(figsize = (8,6))
plt.hist(x, density=True)
sp =np.linspace(0, 50,200)
pdf = stats.norm(x.mean(), x.var()**.5).pdf(sp)
plt.plot(sp, pdf)
plt.ylabel('fraction of samples')
plt.xlabel('worst texture');
```

```python
res = stats.probplot(x, plot=plt)
# есть выбросы по левому и правому хвосту, с -1 по 2 квантили похоже на нормальное распределение,
# т.к. точки лежат на прямой
```

```python
sp =np.linspace(0, 50,200)
cdf = stats.norm(x.mean(), x.var()**.5).cdf(sp)
ecdf = ECDF(x)
plt.figure(figsize = (8,6))
plt.step(ecdf.x, ecdf.y, label='ECDF')
plt.plot(sp,cdf,label='theoretical CDF')
plt.legend(loc='upper left');
```
